## Tensor

Tensors are data structure used by machine learning systems.

Three primary attributes define a tensor:
1. Rank
2. Shape
3. Data type


Rank of tensor is number of indices of tensor. It is also known as "order","degree" or "ndims"


	1) Scalar is not a vector. Its just a number to represent certain quantity without any direction. So it doesn't need any basis vector or Cartesian co-ordinate to represent it. So it is rank 0 or order 0 or 0 degree or 0 dimensional tensor.
	2) Vector needs direction as well as magnitude to represent it. So we need to represent it in certain Cartesian co-ordinate system along with their basis vector. As we always try to represent any object in 3 dimensions( x,y,z), and it required only one index to represent any vector. So it is rank 1 or 1 dimensional tensor.
	3) Matrices needs 2 index to represent it and number of components=3^2=9. Here 3 is dimensions and 2 is number of indexes required to represent that particular tensor. It is also known as rank 2 tensors.
	4) Cube or stack of matrices with shape m×n×c has number of components=3^3=27, because we need 3 indexes are required to represent it. These are known as rank 3 tensors or order 3 tensor.


Shape of tensor is the number of dimensions along each axis.
Example:
A square matrix may have (2, 2) dimensions.
A tensor of rank 3 may have (3, 5, 8) dimensions.

The data type of tensor refers to the type of data contained in it.
Here are some of the supported data types:
	float32,
	float64,
	uint8,
	int32,
	Int64
