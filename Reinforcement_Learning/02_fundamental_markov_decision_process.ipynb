## Markov Decision Processes (MDP)

In case of the k-armed bandit problem, it counts the reward from immediate action which it will take, not the subsequent actions and their rewards.

But MDP, considers sequence of actions and their corresponding rewards. So considering its longterm vision, MDP takes actions which will benefit in longterm. So while making the decision of next action is also considers present state.

## Goal of the reinforcement learning
Main goal of the reinforcement learning is to maximize the expected returns i.e. every action needs to be taken by considering its long-term benefits.<br>

Expected Returns:<br>
$$
E(G_{t}) = E[R_{t+1} + R_{t+2} + R_{t+3} + .... + R_{T}]
$$

Tasks are naturally broken into chunks known as episode. Every episode begins independently of how previous one ended and has its final state which is know as terminal state.  These taks are know as episodic tasks.<br>
