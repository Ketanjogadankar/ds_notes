# Reinforcement Learning

Reinforcement learning is type of machine learning, where model learns from trial and error by exploring all the actions it can take.
Agent is the one who must learn the correct action at time step to maximize the reward.
Goal of the agent is to maximize the reward.

## k-armed bandits

Lets say, doctor has 3 different medicines for same illness, now he is not sure, which will work. So what he will do, he will try all of them one by one and at the end will settle to 1 which has worked for almost all the patients.<br>

Here, doctor is the agent, which decides which medicines to give. So he has to pick 1 from 3 medicines. So here no of actions (k) is 3, i.e. 3 armed bandits. Every action has its associated action value i.e. reward. These values will be decided by algorithm.<br>

Action Value is the reward<br>
$$
q_{*}(a) = \sum_{r} p(r|a)r
$$
The goa of the agent is to maximize the expected reward<br>
$$
\argmax_{a} q_{*}(a)
$$

## Value of an Action

To determine value of an action we use sample average method.
$$
Q_{t}(a) = \frac{\text{Sum of rewards when a taken prior to t}}{\text{number of times action a was taken prior to timestep t}}
$$

i.e. it is the average of the reward received for particular action.<br>


## Selection of Action

Agent selects the action which will return maximum reward. This is the greedy method, it will always select the action with maximum reward, then there will be no exploration even after certain trials. Reward values were given based on sample average method that too on sample time steps only.<br>


## Exploration vs Exploitation

Greedy algorithm always selects the action with the highest reward. This will make model to stick to old values and not to learn or explore further.<br>

There is trade of between exploration and exploitation. Because agent can only do either exploration to find best reward for that action or it can use the rewards value it has learned so far to maximize the reward.<br>

### Epsilon Greedy action selection method
So to solve this issue, we have <b>epsilon greedy method</b>.<br>
Epsilon greedy method has epsilon helps determine how many 1st few steps to be used for exploration. If epsilon is set to 0.01 then 1% of the time steps will be used for exploration and rest for exploitation.<br>



### Optimistic Initial values
Here algorithm assigns maximum reward for each action initially. So agent is optimistic about actions. If that particular action turn out to give correct results then we will give reward 1 and take average to get final reward value for that action. If that action doesn't work, then we will give 0 reward and take average to get final reward.<br>



#### Limitations of optimistic initial value
- optimistic intial values only drive early exploration.
- They are not well-suited for non-stationary problems where, other solution will perform better at different interval.
- We may not know what the optimistic initial value should be


### Upper-Confidence bound(UCB) Action Selection

In UCB, we have upper and lower bound indicating the estimates could vary in those regions only. Higher the gap between upper and lower bound more uncertainty for that particular action.<br>

UCB method relies on uncertaininty of the estimates. Meaning it peaks the value which has highest upper-bound and then updates its confidence interval.
